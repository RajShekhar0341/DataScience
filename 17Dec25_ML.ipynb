# Task
Calculate and visualize the correlation matrix of the `df` DataFrame.

## Calculate Correlation Matrix

### Subtask:
Calculate the pairwise correlation of columns in the `df` DataFrame using `df.corr()`.


**Reasoning**:
Calculate the correlation matrix of the `df` DataFrame using the `.corr()` method, store it in `correlation_matrix`, and display it.



```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

### Visualize Correlation Matrix

#### Subtask:
Create a heatmap of the `correlation_matrix` to visually represent the correlations between variables.

#### Instructions:
1. Import the `seaborn` and `matplotlib.pyplot` libraries.
2. Create a heatmap using `seaborn.heatmap()` with `correlation_matrix` as the data.
3. Add annotations to the heatmap for better readability.
4. Set an appropriate title for the heatmap.
5. Adjust the figure size for better visualization.
6. Display the plot.

**Reasoning**:
To visually represent the correlation matrix, I will create a heatmap using `seaborn` and `matplotlib.pyplot`, incorporating annotations, a title, and an adjusted figure size for clarity.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Wine Quality Dataset')
plt.show()
```
