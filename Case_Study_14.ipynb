Write a Python program to:
• create a set of unique characters from a string
• create a dictionary counting occurrences of each character.

```python
# Input string
text = input("Enter a string: ")

# 1. Create a set of unique characters
unique_chars = set(text)
print("Unique characters in the string:")
print(unique_chars)

# 2. Create a dictionary counting occurrences of each character
char_count = {}
for ch in text:
    char_count[ch] = char_count.get(ch, 0) + 1

print("Character occurrences:")
print(char_count)
```
