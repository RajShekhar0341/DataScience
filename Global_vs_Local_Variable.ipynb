**Write a Python program that defines two functions where one function modifies a variable with the same name as a global variable. Show the difference between local and global scope using the global keyword**

```python
# Global variable
g = 9.8   # acceleration due to gravity (m/s^2)
# Function with local variable (does NOT affect global g)
def local_scope():
    g = 5.0
    print("Inside local_scope(), g =", g)

# Function that modifies the global variable
def global_scope():
    global g
    g = 9.81
    print("Inside global_scope(), g =", g)

# Main program
print("Before function calls, g =", g)

local_scope()
print("After local_scope(), g =", g)

global_scope()
print("After global_scope(), g =", g)
```
