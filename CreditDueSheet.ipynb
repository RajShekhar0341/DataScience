```python
from openpyxl import Workbook, load_workbook
from datetime import datetime
import os

file_name = "credit_due_sheet.xlsx"

# Create file if not exists
if not os.path.exists(file_name):
    wb = Workbook()
    ws = wb.active
    ws.title = "Credit Data"

    # Headers
    ws.append(["Name", "Credit Amount", "Date", "Repaid Amount", "Final Due"])
    wb.save(file_name)

# Load workbook
wb = load_workbook(file_name)
ws = wb.active

def add_entry():
    name = input("Enter Name: ")
    credit = float(input("Enter Credit Amount: "))
    date = input("Enter Date (YYYY-MM-DD): ")
    repaid = float(input("Enter Repaid Amount: "))

    final_due = credit - repaid

    # Append data
    ws.append([name, credit, date, repaid, final_due])
    wb.save(file_name)

    print("✅ Entry added successfully!\n")

def show_data():
    print("\n--- Credit Due Sheet ---")
    for row in ws.iter_rows(values_only=True):
        print(row)
    print()

def menu():
    while True:
        print("1. Add Entry")
        print("2. Show Data")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_entry()
        elif choice == "2":
            show_data()
        elif choice == "3":
            print("Exiting...")
            break
        else:
            print("Invalid choice!\n")

if __name__ == "__main__":
    menu()
```
