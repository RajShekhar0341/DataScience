## 1) Variables, data types, input, and type conversion (implicit & explicit)

```python
# Variables, data types, input, conversions
name = input("Enter your name: ")
age_str = input("Enter your age: ")
height_str = input("Enter your height in meters (e.g., 1.73): ")

# explicit conversions
try:
    age = int(age_str)
except ValueError:
    print("Invalid age; default 0.")
    age = 0

try:
    height = float(height_str)
except ValueError:
    print("Invalid height; default 0.0.")
    height = 0.0

# implicit conversion demo
x = 10      # int
y = 3.5     # float
z = x + y   # implicit int->float

is_adult = age >= 18

print("\n--- Summary ---")
print(f"Name: {name} (type={type(name).__name__})")
print(f"Age: {age} (type={type(age).__name__})")
print(f"Height: {height} (type={type(height).__name__})")
print(f"x + y = {z} (implicit -> {type(z).__name__})")
print(f"Adult? {is_adult} (type={type(is_adult).__name__})")

pi = 3.14159
pi_as_int = int(pi)  # explicit truncation
print(f"Explicit: float {pi} -> int {pi_as_int}")
```

## 2) Operators: Arithmetic, Relational, Logical, Bitwise, Assignment

### 2a) Accept two numbers and perform arithmetic operations

```python
a = float(input("Enter number a: "))
b = float(input("Enter number b: "))

print("\n--- Arithmetic ---")
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
try:
    print(f"a ** b = {a ** b}")
except OverflowError:
    print("a ** b overflowed.")
if b != 0:
    print(f"a / b  = {a / b}")
    print(f"a // b = {a // b}")
    print(f"a % b  = {a % b}")
else:
    print("Division, floor, and modulus undefined for b = 0")
```

### 2b) Logical condition: check if a number is positive and even

```python
n = int(input("Enter an integer: "))
is_positive = n > 0
is_even = (n % 2 == 0)
if is_positive and is_even:
    print(f"{n} is positive and even.")
else:
    print(f"{n} is NOT both positive and even.")
```

```python
x = int(input("Enter x: "))
y = int(input("Enter y: "))
print(f"Before swap: x={x}, y={y}")
x ^= y
y ^= x
x ^= y
print(f"After  swap: x={x}, y={y}")
```

## 4) Loops: for, while, range

### 4a) First n Fibonacci numbers (while loop)

```python
n = int(input("Enter n (>=1) for Fibonacci: "))
a, b = 0, 1
count = 0
fib_list = []
while count < n:
    fib_list.append(a)
    a, b = b, a + b
    count += 1
print("Fibonacci:", fib_list)
```

### 4b) Sum of 1..n using for loop

```python
n = int(input("Enter n (>=1) to sum 1..n: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum =", total)
```

### 4c) Factorial using while loop

```python
n = int(input("Enter n (>=0) for factorial: "))
if n < 0:
    print("Factorial not defined for negative numbers.")
else:
    fact = 1
    i = 1
    while i <= n:
        fact *= i
        i += 1
    print(f"{n}! = {fact}")
```

## 5) Nested loops: patterns

### 5a) Inverted pyramid pattern

```python
rows = int(input("Enter number of rows for inverted pyramid: "))
for i in range(rows):
    spaces = " " * i
    stars = "*" * (2 * (rows - i) - 1)
    print(spaces + stars)
```

### 5b) Floyd’s Triangle

```python
rows = int(input("Enter number of rows for Floyd's Triangle: "))
num = 1
for i in range(1, rows + 1):
    line = []
    for _ in range(i):
        line.append(str(num))
        num += 1
    print(" ".join(line))
```

## 7) User-defined functions, scope/lifetime, and return values

```python
def factorial(n: int) -> int:
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print("5! =", factorial(5))
```

```python
counter = 0  # global variable

def increment_local(counter):
    counter += 1
    return counter

def increment_global():
    global counter
    counter += 1
    return counter

def outer():
    msg = "outer"
    def inner():
        nonlocal msg
        msg = "updated by inner"
        return msg
    inner()
    return msg

print("increment_local(10) ->", increment_local(10))
print("global before ->", counter)
print("increment_global() ->", increment_global())
print("global after  ->", counter)
print("outer()/nonlocal demo ->", outer())
```

## 8) Prime checker function

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    if n % 2 == 0:
        return n == 2
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

for x in [0, 1, 2, 3, 4, 17, 18, 19, 20]:
    print(f"{x} prime? -> {is_prime(x)}")
```

## 11) Lists: indexing, slicing, modify, append/insert/remove + Nested lists

```python
students = ["Raj", "Swathi", "Amala", "Gulshan", "Anudeep"]
print("First:", students[0])
print("Last :", students[-1])
print("Slice [1:4]:", students[1:4])

students[2] = "Amala R"
print("After modify:", students)

students.append("Kalpana")
print("After append:", students)

students.insert(2, "Vishal")
print("After insert:", students)

name = "Gulshan"
if name in students:
    students.remove(name)
    print(f"After remove({name!r}):", students)

popped = students.pop()
print("Popped:", popped)
print("Final list:", students)

records = [
    ["Raj",     [85, 90, 88]],
    ["Swathi",  [92, 81, 95]],
    ["Amala",   [78, 84, 80]],
]
print("Amala 2nd mark:", records[2][1][1])
records[0][1].append(93)
print("Raj marks after append:", records[0][1])
for name, marks in records:
    avg = sum(marks) / len(marks)
    print(f"{name}: marks={marks}, avg={avg:.2f}")
```

## 12) Stack using list: push and pop

```python
stack = []

def push(item):
    stack.append(item)
    print(f"Pushed {item}; stack -> {stack}")

def pop_():
    if not stack:
        print("Underflow: stack is empty.")
        return None
    item = stack.pop()
    print(f"Popped {item}; stack -> {stack}")
    return item

push(10)
push(20)
push(30)
pop_()
pop_()
pop_()
pop_()  # underflow
```
