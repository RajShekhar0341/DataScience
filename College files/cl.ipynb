```python
def is_palindrome_number(n):
    """
    Check if the given integer n is a palindrome number.

    Args:
        n (int): The number to check.

    Returns:
        bool: True if n is a palindrome, False otherwise.
    """
    original = str(n)
    reversed_num = original[::-1]
    return original == reversed_num

# Example usage:
number = 121
print(f"{number} is palindrome: {is_palindrome_number(number)}")
```
