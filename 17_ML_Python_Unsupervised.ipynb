```python
import pandas as pd
df = pd.read_excel('/content/call_center_kmeans_50.csv')
df.head()
```

```python
from sklearn.cluster import KMeans
for i in range(1,10):
  kms = KMeans(n_clusters=i)
  kms.fit(df)
  losses.append(kms.inertia_)
  print(losses)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,10),losses[:len(range(1,10))])
plt.show()
```

```python
plt.scatter(df['Experience_in_months'], df['Calls_Attended_per_day'])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmodel=KMeans(n_clusters=3)
kmodel.fit(df)
```
