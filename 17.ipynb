```python
#PCA STEP 1: IMPORT Aand load data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#load dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(X)
```

```python
#step 2: Standardize the data
#Mean Centering
X_meaned = X - np.mean(X, axis=0)
print(X_meaned)
```

```python
#step 3: compute covariance matrix
cov_matrix = np.cov(X_meaned, rowvar=False)
print(cov_matrix)
```

```python
#STEP 4: compute eigenvalues & eigenvectors
eigen_values, eigenvectors = np.linalg.eigh(cov_matrix)
print(eigen_values)
print(eigenvectors)
```

```python
#STEP 5: Sort EigenValues (Descending)
sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalues = eigen_values[sorted_index]
sorted_eigenvectors = eigenvectors[:,sorted_index]
print(sorted_eigenvalues)
print(sorted_eigenvectors)
```

```python
from IPython.utils.py3compat import no_code
#step 6: slect top k components (k=2)
no_components = 2
eigenvector_subset = sorted_eigenvectors[:,0:no_components]
print(eigenvector_subset)
```

```python
#step 7: Transform data
X_pca = np.dot(eigenvector_subset.transpose(),X_meaned.transpose()).transpose()
print(X_pca)
#
```

```python
#step 8: visualization
plt.figure()
for i in np.unique(y):
  plt.scatter(X_pca[y==i,0],X_pca[y==i,1],label=i)
  plt.title("PCA - Manual Implementation")
  plt.xlabel("PC1")
  plt.ylabel("PC2")
  plt.legend()
  plt.show()
```
