```python
import numpy as np
x=np.array([4,5.6,7,9.2,8])
y=np.array([5,7.6,8,9.3,7.8])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
w0=5
w1=4
yh=w0 + w1 * x
print(yh)
```

```python
plt.scatter(x,y)
plt.plot(x,yh)
plt.show()
```

```python


mean_squared_error=np.mean((yh-y)**2)
print(mean_squared_error)
```

```python
dew0 = -2*(y-yh).mean()
dew1 = -2*((y-yh)*x).mean()
print(dew0,dew1)
```

```python
lr = 0.01
w0 = w0-lr * dew0
w1 = w1-lr * dew1
print(w0,w1)
```

```python
yh = w0 + w1 * x
print(yh)
```

```python
print(w0,w1)
```

```python
plt.scatter(x,y)
plt.plot(x,yh, 'r')
plt.show()
```
