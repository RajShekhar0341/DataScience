````python
```python
def fibonacci_recursive(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    else:
        series = fibonacci_recursive(n - 1)
        if n == 2:
            series.append(1)
        else:
            series.append(series[-1] + series[-2])
        return series

def print_fibonacci_series():
    try:
        num_elements = int(input("Enter the number of elements for the Fibonacci series: "))
        if num_elements < 0:
            print("Please enter a non-negative integer.")
        else:
            series = fibonacci_recursive(num_elements)
            print(f"Fibonacci series of {num_elements} elements: {series}")
    except ValueError:
        print("Invalid input. Please enter an integer.")

# Call the function to print the Fibonacci series
print_fibonacci_series()
```
````

````python
# Recursive function to find nth Fibonacci number
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Input number of elements
n = int(input("Enter number of elements: "))

print("Fibonacci series:")
for i in range(n):
    print(fibonacci(i), end=" ")
````

````python
# Recursive function to find nth Fibonacci number
def fibonacci(n):
  if n <= 1:
      return n
  else:
      return fibonacci(n-1) + fibonacci(n-2)

#input number of elements
n = int(input("Enter number of elements: "))

print("Fibonacci series:")
for i in range(n):
  print(fibonacci(i), end=" ")
````
