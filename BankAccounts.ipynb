```python
# Fixed bank account names
banks = [
    "SBI Gulshan",
    "HDFC Madhusudha",
    "Harsh Construction",
    "Union Bank Harsha",
    "SBI Chakri",
    "Axis Swathi",
    "Axis Raj"
]

# Input date
date = input("Enter date (YYYY-MM-DD): ")

balances = []

print("\nEnter balances for each bank:\n")

# Input balances
for bank in banks:
    bal = float(input(f"{bank}: "))
    balances.append(bal)

# Calculate total
total_balance = sum(balances)

# Display output
print("\n--- Bank Balances ---")
print("Date:", date)

for i in range(len(banks)):
    print(f"{banks[i]}: {balances[i]}")

print("\nTotal Balance:", total_balance)
```

```python
import pandas as pd

# Create a dictionary for the data
data = {
    'Bank Name': banks,
    'Balance': balances
}

# Create a DataFrame
df = pd.DataFrame(data)

# Add the date as a separate column (if you want it on each row) or as metadata
# For simplicity, let's add it as a column for now
df['Date'] = date

# Reorder columns to have Date first, then Bank Name, then Balance
df = df[['Date', 'Bank Name', 'Balance']]

# Save to Excel
excel_filename = 'bank_balances.xlsx'
df.to_excel(excel_filename, index=False)

print(f"Bank balances saved to {excel_filename}")
```
