#2.Built in functions to edit the string

```python
s = "welcome to GITAM"
new_s = s.replace("w", "W", 1)
#Number 1 indicates first occurrence
print(new_s)
print(s.title())
print(s.swapcase())
print("Split by space -->", s.split())
print("Count 'l'  -->", s.count('l'))
s2 = "   Python is fun!   "
print("Original with spaces:", repr(s2))
print("After strip()       :", repr(s2.strip()))
print("After lstrip()      :", repr(s2.lstrip()))
print("After rstrip()      :", repr(s2.rstrip()))
```

#String Slicing

```python
print("\nString Slicing")
university_name="GITAM University"
print(university_name[0:5])
print(university_name[::-1])
print(university_name[-5:-3])
print(university_name[6:])
print(university_name[:5])
print(university_name[0:8:3])
```
