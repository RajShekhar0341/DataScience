```python
name = input("Enter name: ")
credit = float(input("Enter credit amount: "))
debit = float(input("Enter debit amount: "))
repaid = float(input("Enter repaid amount: "))
date = input("Enter date (e.g., YYYY-MM-DD): ")
balance = float(input("Enter balance: "))

print("\n--- Entered Information ---")
print(f"Name: {name}")
print(f"Credit: {credit}")
print(f"Debit: {debit}")
print(f"Repaid: {repaid}")
print(f"Date: {date}")
print(f"Balance: {balance}")
```
